```python
import ipywidgets as iw
import PIL
from fastai import vision
from fastbook import *
```

```python
upload_botton = iw.widgets.FileUpload(multiple = False)
upload_botton
```

```python
image = PILImage.create(upload_botton.data[-1])
```

```python
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: 
  display(image.to_thumb(256,256))
out_pl
```

```python
learn_function = load_learner('/content/drive/MyDrive/models/export.pkl')
```

```python
prediction,prediction_index,probability = learn_function.predict(image)
```

```python
prediction, prediction_index, probability
```

```python
prediction_label = iw.widgets.Label()
prediction_label.value = f'Prediction: {prediction}; Probability: {probability[prediction_index]:.04f}'
prediction_label
```

```python
operate_botton = iw.widgets.Button(description='Classify')
operate_botton
```

```python
def on_click_classify(change):
    img = PILImage.create(upload_botton.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    prediction,prediction_index,probability = learn_function.predict(img)
    prediction_label.value = f'Prediction: {prediction}; Probability: {probability[prediction_index]:.04f}'

operate_botton.on_click(on_click_classify)
```

```python
upload_botton = widgets.FileUpload()
```

```python
iw.VBox([widgets.Label('Grab Your Guitar For Detection'), 
      upload_botton, operate_botton, out_pl, prediction_label])
```
